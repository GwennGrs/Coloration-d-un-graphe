## Nombre de couleurs
On peut alors avoir 2 couleurs : une pour le centre et une pour tous les autres points différent du centre de l'étoile

```python
def graph_etoile(n):
    G = nx.Graph()
    G.add_node(0)
    for i in range(1, n):
        G.add_node(i)
        G.add_edge(0, i)
    return G
```

```python
A = graph_etoile(9)
nx.draw_networkx(A)
plt.show()
```

```python
def color_graph_etoile(G):
    n = G.number_of_nodes()
    l = []
    l.append("red")
    for i in range(1,n):
        l.append("blue")
    return l
l = color_graph_etoile(A)
nx.draw_networkx(A,node_color=l)
```
