# Graphe complet
[Un graphe complet](https://fr.wikipedia.org/wiki/Graphe_complet) est un graphe où tout les sommets sont reliés.

## Exemple

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import queue
def graphe_complet(n):
    G = nx.complete_graph(n)
    nx.draw_networkx(G)
    plt.show()
    return G

graphe_complet(5)
```

## Coloration du graphe
On en vient rapidement à la conclusion que l'on à besoin de n couleurs, puisque tous les noeuds sont voisins.

```python
def color_complet_graphe(G):
    colors = plt.cm.rainbow(np.linspace(0, 1, len(G.nodes)))
    nx.draw_networkx(G, pos=nx.spring_layout(G), node_color=colors, with_labels=True)
    plt.show()

G = nx.complete_graph(5)
color_complet_graphe(G)
```
