# Greedy-Color de NetworkX

Il utilise un algorithme glouton: il attribue une couleur au nœud courant en choisissant la plus petite couleur disponible (non utilisée par ses voisins). Cela se répète pour tous les nœuds dans un ordre donné. 
## Questions 2 - valadité

Cet algorithme est valide mais non optimal. Il existe plusieurs cas où le nombre de couleurs est au dessus du nombres chromatique. Notamment pour des graphes précis comme les graphes bipartis, cycles ou encore les graphes complets. La solution sera bonne, mais pas la plus optimal. La qualité dépend souvent de l'ordre de traitement des noeuds qui peut minimiser le résultât, il utilise un ordre déterminé par des stratégies comme largest_first, random_sequential, ou saturation_largest_first. Il produit toujours une coloration valide, mais ne garantit pas le nombre chromatique minimal, car le nombre de couleurs dépend de l'ordre des sommets. Par exemple, pour un cycle (nombre chromatique 2 si pair, 3 si impair), largest_first et saturation_largest_first donnent souvent des résultats proches de l'optimum, tandis que random_sequential peut utiliser plus de couleurs.[Wikipédia](https://en.wikipedia.org/wiki/Greedy_coloring), [Documentation networkx](https://networkx.org/documentation/stable/auto_examples/algorithms/plot_greedy_coloring.html)

## Question 4 - Visualisation de l'algorithme 

Le nombre chromatique d'un graphe de Petersen [est de 3](https://fr.wikipedia.org/wiki/Graphe_de_Petersen#:~:text=Il%20existe%20donc%20une%204,et%20d'indice%20chromatique%204.). On peut voir avec cet exemple que lors du lancement avec l'algorithme de greey_color, nous trouvons au dessus de son nombre chromatique.

## Complexité

La complexité varie : O(n log n + m) pour largest_first (tri des degrés), O(n + m) pour random_sequential (ordre aléatoire), et jusqu'à O(n^2 log n) pour saturation_largest_first (gestion des saturations)

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.petersen_graph()

coloring = nx.coloring.greedy_color(G, strategy='random_sequential')
num_colors = len(set(coloring.values()))
print(f"Coloration : {coloring}")
print(f"Nombre de couleurs : {num_colors}")

color_map = [coloring[node] for node in G.nodes()]
nx.draw(G, nx.spring_layout(G), node_color=color_map, with_labels=True, cmap=plt.cm.Set3)
plt.title("Coloration Petersen (random_sequential)")
plt.show()
```
