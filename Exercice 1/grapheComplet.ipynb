```python
%matplotlib inline
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import queue
```

```python
def graphe_complet(n):
    G = nx.complete_graph(n)
    nx.draw_networkx(G)
    plt.show()
    return G

graphe_complet(5)
```

```python
def color_complet_graphe(G):
    pos = nx.spring_layout(G)
    colors = plt.cm.rainbow(np.linspace(0, 1, len(G.nodes)))
    nx.draw_networkx(G, node_color=colors, with_labels=True)
    plt.show()
    return colors

color_complet_graphe(graphe_complet(5))
```
