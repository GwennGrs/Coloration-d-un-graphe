On devine facilement que pour un graphe cyclique le nombre de couleur est 2 si n est pair et 3 si n est impair.

```python
#Exercice 1
import matplotlib.pyplot as plt
import networkx as nx

def genere_graphe_cycle(n):
    """Génère un graphe cycle de n sommets."""
    G = nx.cycle_graph(n)
    return G

def colorie_graphe(G):
    """Colorie les sommets du graphe en alternant les couleurs."""
    n = len(G.nodes)
    if n % 2 == 0:
        colors = ['red' if i % 2 == 0 else 'blue' for i in range(n)]
    else:
        colors = ['red' if i % 3 == 0 else 'blue' if i % 3 == 1 else 'green' for i in range(n)]
    return colors

G = genere_graphe_cycle(16)
colors = colorie_graphe(G)
nx.draw_networkx(G, node_color=colors, with_labels=True)
plt.show()


```
