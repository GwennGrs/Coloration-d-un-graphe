On devine facilement que pour un graphe cyclique le nombre de couleur est 2 si n est pair et 3 si n est impair.

```python
#Exercice 1
import matplotlib.pyplot as plt
import networkx as nx

def genere_graphe_cycle(n):
    """Génère un graphe cycle de n sommets."""
    G = nx.cycle_graph(n)
    return G

def colorie_graphe(G):
    """Colorie les sommets du graphe en alternant les couleurs."""
    n = len(G.nodes)
    if n % 2 == 0:
        colors = ['red' if i % 2 == 0 else 'blue' for i in range(n)]
    else:
        colors = ['red' if i % 3 == 0 else 'blue' if i % 3 == 1 else 'green' for i in range(n)]
    return colors

G = genere_graphe_cycle(21)
colors = colorie_graphe(G)
nx.draw_networkx(G, node_color=colors, with_labels=True)
plt.show()


```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Création du graphe
europe = nx.Graph()

countries = [
    'France', 'Germany', 'Italy', 'Spain', 'Belgium', 'Netherlands', 'Switzerland', 'Austria',
    'Portugal', 'United Kingdom', 'Ireland', 'Denmark', 'Norway', 'Sweden', 'Finland',
    'Poland', 'Czech Republic', 'Slovakia', 'Hungary', 'Romania', 'Bulgaria', 'Greece',
    'Croatia', 'Slovenia', 'Serbia', 'Bosnia and Herzegovina', 'Albania', 'North Macedonia',
    'Montenegro', 'Luxembourg'
]

europe.add_nodes_from(countries)

borders = [
    ('France', 'Germany'), ('France', 'Italy'), ('France', 'Spain'), ('France', 'Belgium'), ('France', 'Luxembourg'),
    ('France', 'Switzerland'), ('Germany', 'Belgium'), ('Germany', 'Netherlands'), ('Germany', 'Switzerland'),
    ('Germany', 'Austria'), ('Germany', 'Poland'), ('Germany', 'Czech Republic'), ('Germany', 'Luxembourg'),
    ('Italy', 'Switzerland'), ('Italy', 'Austria'), ('Italy', 'Slovenia'), ('Spain', 'Portugal'),
    ('Belgium', 'Netherlands'), ('Belgium', 'Luxembourg'), ('Netherlands', 'Germany'), ('Switzerland', 'Austria'),
    ('Austria', 'Czech Republic'), ('Austria', 'Slovakia'), ('Austria', 'Hungary'), ('Austria', 'Slovenia'),
    ('Portugal', 'Spain'), ('United Kingdom', 'Ireland'), ('Denmark', 'Germany'), ('Norway', 'Sweden'),
    ('Sweden', 'Finland'), ('Poland', 'Czech Republic'), ('Poland', 'Slovakia'), ('Poland', 'Germany'),
    ('Czech Republic', 'Slovakia'), ('Slovakia', 'Hungary'), ('Hungary', 'Romania'), ('Hungary', 'Serbia'),
    ('Hungary', 'Croatia'), ('Hungary', 'Slovenia'), ('Romania', 'Bulgaria'), ('Romania', 'Serbia'),
    ('Bulgaria', 'Greece'), ('Bulgaria', 'Serbia'), ('Greece', 'Albania'), ('Greece', 'North Macedonia'),
    ('Croatia', 'Slovenia'), ('Croatia', 'Bosnia and Herzegovina'), ('Croatia', 'Serbia'), ('Croatia', 'Montenegro'),
    ('Slovenia', 'Croatia'), ('Serbia', 'Bosnia and Herzegovina'), ('Serbia', 'Montenegro'),
    ('Serbia', 'North Macedonia'), ('Serbia', 'Albania'), ('Montenegro', 'Albania'), ('Albania', 'North Macedonia')
]

europe.add_edges_from(borders)

pos = nx.spring_layout(europe, k=0.5, iterations=50) 

nx.draw_networkx(
    europe, 
    pos=pos, 
    with_labels=True, 
    node_color='lightblue', 
    edge_color='gray', 
    width=2.0,  
    node_size=250, 
    font_size=10
)

plt.title("Graphe simplifié de l'Europe (version étendue)")
plt.axis('off')  
plt.show()
```
